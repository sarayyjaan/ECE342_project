```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
columns = ["time", "x_axis", "y_axis", "z_axis"]
df = pd.read_csv("/content/gdrive/MyDrive/stationarydata.csv", usecols=columns)
df_walk = pd.read_csv("/content/gdrive/MyDrive/walk.csv", usecols=columns)
```

```python
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
#print("Contents in csv file:", df)
df.plot('time', y=['x_axis', 'y_axis', 'z_axis'])
df_walk.plot('time', y=['x_axis', 'y_axis', 'z_axis'])
plt.show()
```
